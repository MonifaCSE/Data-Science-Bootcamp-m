```python
s = Service('chromedriver')
driver = webdriver.Chrome(service=s)
# session = HTMLSession()
start = time.time()
 
all_urls = []
url_cnt = 0
 
airbnb_url ='https://www.airbnb.com/?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&search_mode=flex_destinations_search&flexible_trip_lengths%5B%5D=one_week&location_search=MIN_MAP_BOUNDS&price_filter_input_type=0&category_tag=Tag%3A7769&search_type=category_change'
print(airbnb_url)

driver.get(airbnb_url)
time.sleep(3)
product_link = driver.find_elements(by=By.CLASS_NAME, value='ln2bl2p.dir.dir-ltr')
product_urls = [p.get_attribute('href') for p in product_link]
for p in product_urls:
    all_urls.append(p)
    url_cnt += 1
    
    print(" URL: " + str(url_cnt))
    print(p)
        
#     clear_output(wait=True)
print(time.time()-start)
# driver.maximize_window()
```

```python
with open('beach_product_urls.txt', 'w') as f:
    for s in all_urls:
        f.write(s)
        f.write('\n')
```

```python
def title_scrap(driver):
    try:
        title = driver.find_element(by=By.CLASS_NAME, value='_fecoyn4').text
        print("Title: ", title)
        return title        
    except:
        title="Beach Not Found!"
        print(title)
        return None
```

```python
def rating_scrap(driver):
    try:
        rating = driver.find_element(by=By.CLASS_NAME, value='_17p6nbba').text
        print("Ratings: ", rating)

        return rating
    except:
        rating="Ratings Not Found!"
        print(rating)
        return None
```

```python
def reviews_scrap(driver):
    try:
        reviews = driver.find_element(by=By.CLASS_NAME, value='_s65ijh7').text
        print("Reviews: ", reviews)

        return reviews
    except:
        reviews="Reviews Not Found!"
        print(reviews)
        return None
```

```python
def location_scrap(driver):
    try:
        location = driver.find_element(by=By.CLASS_NAME, value='_9xiloll').text
        print("LOcation: ", location)

        return location
    except:
        location="Location Not Found!"
        print(location)
        return None
```

```python
def price_scrap(driver):
    try:
        price = driver.find_element(by=By.XPATH, value='//*[@id="site-content"]/div/div[1]/div[3]/div/div[2]/div/div/div[1]/div/div/div/div/div/div/div[1]/div[1]/div[1]/div/span/div').text
        print("Price: ", price)

        return price
    except:
        price="Price Not Found!"
        print(price)
        return None
```

```python
def place_offer_scrap(driver):
    try:
        place_offer = driver.find_element(by=By.CLASS_NAME, value='_1byskwn').text.split('\n')
        print("Place Offer: ", place_offer)

        return place_offer
    except:
        place_offer="Place Offer Not Found!"
        print(place_offer)
        return None
```

```python
def image_scrap(driver, category):
    try:
        images = driver.find_element(by=By.CLASS_NAME, value='_skzmvy')
        img_path = []
        for img in images.find_elements(by=By.TAG_NAME,value='img'):
            img_urls = img.get_attribute('src')

            img_urls_rename = '/'.join(img_urls.split('?')[:-1])
             
            img_name = title+'@'.join(img_urls_rename.split('/')[-2:])

            r = requests.get(img_urls_rename)
            i = Image.open(BytesIO(r.content))
            i.save(f"img/{category}/{img_name}.jpg")
            img_path.append(img_name)
        return img_path
    except:
        print('No images')        
        return 'N/A'
```

```python
def meta_data_scrap(driver):
    try:
        meta_data = driver.find_element(by=By.XPATH, value='//*[@id="site-content"]/div/div[1]/div[3]/div/div[1]/div/div[4]/div/div[2]')

        md = meta_data.find_elements(by=By.TAG_NAME,value='div')
        meta_dic = {}
        for m in md:
            m = m.text
            if '\n' in m:
                key, val = m.split('\n')[0].replace(':', ''), m.split('\n')[1]
                meta_dic[key] = val
        print("Description: ", meta_dic)
        return meta_dic
    except:
        meta_data='No meta data found'
        print(meta_data)        
        return 'N/A'
```

```python
def show_more_click(driver):    
    try:
        driver.find_element(by=By.CLASS_NAME, value='cn5lml1.dir.dir-ltr').click().alert().getText()
#         s_driver.find_element(by=By.XPATH, value='//*[@id="app-wrapper"]/div[1]/div[2]/div[2]/div[3]/div[2]/div/div[1]/div/div[2]/div[3]/div/div[2]/div/div[2]/button').click()
        time.sleep(2)
    except:
        
        print("No 'Show more' button")
```

```python
def open_driver():
    op = webdriver.ChromeOptions()
    op.add_argument('headless')
    s = Service('chromedriver')
    driver = webdriver.Chrome(service=s, options=op)
    return driver
```

```python
def product_scrap(driver, url, category):
    driver.get(url)
    time.sleep(5)

    # single page data scrapping
    title = title_scrap(driver)
    # if title not found (product sold) don't proceed further for other data
    rating = rating_scrap(driver)
    
    reviews= reviews_scrap(driver)
    
     
    location = location_scrap(driver)
    
    # scrapping price
    price = price_scrap(driver)
    
    place_offer=place_offer_scrap(driver)
    
    # scrapping all images
    images = image_scrap(driver, category)
    

    # extractig meta data
    meta_data = meta_data_scrap(driver)
        
    # click on show_more button
    show_more_click(driver)

     
    
    return title, rating,reviews, location,price,place_offer, images, meta_data
```

```python
category = 'beach'
with open('beach_product_urls.txt', 'r') as f:
    single_url = f.readlines()
    titles, ratings,reviewss, locations,prices,place_offers, img_lst,  meta_lst = [], [], [], [], [], [], [],[]
    cnt = 1
#     print(single_url)
    for s in single_url:
        print(cnt, s)
        driver = open_driver()
        
        title, rating,reviews, location,price,place_offer, images, meta_data = product_scrap(driver, s, category)
        
        titles.append(title)
        ratings.append(rating)
        reviewss.append(reviews)
        locations.append(location)
        prices.append(price)
        place_offers.append(place_offer)
        img_lst.append(images)
        meta_lst.append(meta_data)


        cnt += 1
        print("=============================================")
        
#with open(f"beach.txt","a") as f:
#  f.write("Titles,   Ratings,   Reviews  ,locations  ,Prices   ,place_offers, Images , Description")
#  f.write('\n')
#  for t,r,re,l,p,o,i,d  in zip(titles, ratings,reviewss, locations,prices,place_offers, img_lst, meta_lst):

#      f.write(f"{t},  {r},  {re},  {l},  {p},  {o}, {i}, {d}")
#      f.write('\n')
```
